```python
import torch

# Giả sử bạn có file model.pt
path_to_model = './model/text_cnn_model.pt'

# Tải file lên
# weights_only=True là tiêu chuẩn an toàn mới để tránh thực thi code độc hại
checkpoint = torch.load(path_to_model, map_location='mps')

# In ra danh sách các lớp và kích thước ma trận trọng số
for layer_name, weights in checkpoint.items():
    print(f"Lớp: {layer_name} | Kích thước ma trận: {weights.shape}")
```
