```python
import numpy as np
```

## ME

```python
actual = [3, 5, 2, 8, 7]
predicted1 = [2.5, 5.3, 1.9, 7.8, 7.5]
predicted2 = [3.5, 5.3, 2.1, 8.2, 7.5]
predicted3 = [2.5, 4.7, 1.9, 7.8, 6.5]
```

```python
def me(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    errors = actual - predicted
    return np.mean(errors)
```

```python
print(me(actual, predicted1))
print(me(actual, predicted2))
print(me(actual, predicted3))
```

## MAE

```python
actual = [3, 5, 2, 8, 7]
predicted1 = [2.5, 5.3, 1.9, 7.8, 7]
predicted2 = [3.5, 5.3, 2.1, 8.2, 7.5]
predicted3 = [2.5, 4.7, 1.9, 7.8, 6.5]
```

```python
def mae(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    errors = np.abs(actual - predicted)
    return np.mean(errors)
```

```python
print(mae(actual, predicted1))
print(mae(actual, predicted2))
print(mae(actual, predicted3))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(actual, predicted1))
print(mean_absolute_error(actual, predicted2))
print(mean_absolute_error(actual, predicted3))
```

## MSE

```python
actual = [3, 5, 2, 8, 7]
predicted1 = [2.5, 5.3, 1.9, 7.8, 7]
predicted2 = [3.5, 5.3, 2.1, 8.2, 7.5]
predicted3 = [2.5, 4.7, 1.9, 7.8, 6.5]
```

```python
def mse(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    errors = (actual - predicted) ** 2
    return np.mean(errors)
```

```python
print(mse(actual, predicted1))
print(mse(actual, predicted2))
print(mse(actual, predicted3))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(actual, predicted1))
print(mean_squared_error(actual, predicted2))
print(mean_squared_error(actual, predicted3))
```

## MAPE

```python
actual = [3, 5, 2, 8, 7]
predicted1 = [2, 2, 2, 2, 2]
predicted2 = [5, 5, 5, 5, 5]
predicted3 = [7, 7, 7, 7, 7]
```

```python
def mape(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    percentage_errors = np.abs((actual - predicted) / actual) * 100
    return np.mean(percentage_errors)
```

```python
print(mape(actual, predicted1))
print(mape(actual, predicted2))
print(mape(actual, predicted3))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
print(mean_absolute_percentage_error(actual, predicted1))
print(mean_absolute_percentage_error(actual, predicted2))
print(mean_absolute_percentage_error(actual, predicted3))
```

```python
actual = [3, 5, 2, 8, 7]
predicted1 = [6, 10, 4, 16, 14]
predicted2 = [9, 15, 6, 24, 21]
predicted3 = [12, 20, 8, 32, 28]


print(mape(actual, predicted1))
print(mape(actual, predicted2))
print(mape(actual, predicted3))

print(mean_absolute_percentage_error(actual, predicted1))
print(mean_absolute_percentage_error(actual, predicted2))
print(mean_absolute_percentage_error(actual, predicted3))
```

## sMAPE

```python
actual = np.array([3, 5, 2, 8, 7])
predicted1 = np.array([2, 2, 2, 2, 2])
predicted2 = np.array([5, 5, 5, 5, 5])
predicted3 = np.array([7, 7, 7, 7, 7])
```

```python
def smape(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    numerator = np.abs(actual - predicted)
    denominator = np.abs(actual) + np.abs(predicted)
    smape_values = numerator / denominator
    return np.mean(smape_values) * 200
```

```python
print(smape(actual, predicted1))
print(smape(actual, predicted2))
print(smape(actual, predicted3))
```

```python
from sktime.performance_metrics.forecasting import MeanAbsolutePercentageError
```

```python
print(MeanAbsolutePercentageError(symmetric=True)(actual, predicted1))
print(MeanAbsolutePercentageError(symmetric=True)(actual, predicted2))
print(MeanAbsolutePercentageError(symmetric=True)(actual, predicted3))
```
