```python
import numpy as np
```

```python
z = np.array([2.0, 1.0, 0.1])
def softmax(z):
    exp_z = np.exp(z)
    return exp_z / sum(exp_z)
print(softmax(z))
```

```python
temperature = 2
def softmax_temperature(z, temperature):
    exp_z_t = np.exp(z/temperature)
    return exp_z_t/sum(exp_z_t)

print(softmax_temperature(z, temperature))
```

```python
z = np.array([2, 0])
print(softmax(z))
print(softmax(z/2))
```

```python
z = np.array([3, 1, 0])
exp_z = np.exp(z)
print(exp_z)
print(sum(exp_z))
print(exp_z/sum(exp_z))
```

```python
print(softmax([1.0, 1.0, 1.0]))
```

```python
print(softmax_temperature(np.array([2,1,0]), 10))
```
