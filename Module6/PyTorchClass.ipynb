```python
import torch
import torch.nn as nn
```

```python
# Thay vì tạo mô hình bằng nn.Sequential thì ta có thể tự định nghĩa bằng cách này
class MyMLP(torch.nn.Module):
    def __init__(self, hidden_data):
        super(MyMLP, self).__init__()
        self.fc1 = nn.Linear(784, 256)
        self.activation = nn.ReLU()
        self.fc2 = nn.Linear(256, 10)

    def forward(self, x):
        x = self.fc1(x)
        x = self.activation(x)
        x = self.fc2(x)
        return x


model = MyMLP(256)
X = torch.rand(32, 784)

output = model(X)
print(output.shape)
```
