```python
import numpy as np
```

```python
def predict(X, theta):
    z = np.dot(X, theta)
    return 1 / (1 + np.exp(-z))


X = [[-3.2, 0.5, 2.0, 1]]
theta = [0.4, -0.3, 0.2, -0.1]

print(predict(X, theta))
```

```python
def compute_loss(y_hat, y):
    y_hat = np.clip(y_hat, 1e-7, 1 - 1e-7)
    return (-y * np.log(y_hat) - (1 - y) * np.log(1 - y_hat)).mean()


y = np.array([1, 1, 0, 0])
y_hat = np.array([0.9, 0.6, 0.2, 0.4])

print(compute_loss(y_hat, y))
```

```python
def compute_gradient(X, y_true, y_pred):
    gradient = np.dot(X.T, (y_pred - y_true)) / y_true.size
    return gradient


X = np.array([[1, 3], [2, 1], [3, 2], [1, 2]])
y_true = np.array([1, 0, 1, 0])
y_pred = np.array([0.7, 0.3, 0.8, 0.2])

print(compute_gradient(X, y_true, y_pred))
```

```python
def compute_accuracy(y_true, y_pred):
    y_pred_rounded = np.round(y_pred)
    accuracy = np.mean(y_true == y_pred_rounded)
    return accuracy


y_true = np.array([1, 0, 1, 0, 1])
y_pred = np.array([0.84, 0.49, 0.4, 0.2, 0.77])

print(compute_accuracy(y_true, y_pred))
```
