```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

tickers = ["VCB.VN", "FPT.VN", "HPG.VN"]
start_date = "2020-01-01"
end_date = "2026-02-15"
```

```python
for ticker in tickers:
    df = yf.download(ticker, start=start_date, end=end_date)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df = df.reset_index()
    df.columns.values[0] = 'Date' # Cưỡng ép cột đầu tiên tên là Date

    df['Date'] = pd.to_datetime(df['Date'])
    for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    fig = go.Figure(data=[go.Candlestick(
        x=df['Date'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close']
    )])

    fig.update_layout(title=f'Dữ liệu chuẩn mã {ticker}', template='plotly_dark')
    pio.write_html(fig, file=f'ket_qua_{ticker}.html', auto_open=True)
    df.to_csv(f"{ticker}.csv")
```
